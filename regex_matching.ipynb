```python
class Solution:
    def isMatch(self, s, p):
        """
        :type s: str
        :type p: str
        :rtype: bool
        """
        boolean = True
        i = 0
        j = 0
        while i < len(list(s)) and boolean:
            try: # create the candidate for match.
                if p[j] == '*':
                    if s[i] == s[i-1] or p[j-1] == '.': # if there is *: take the previous letter as a candidate.
                        candidate = p[j-1]
                    else:
                        candidate = p[j+1] # case where it ends *.
                        if candidate == p[j-1]:
                            j += 3
                        else:
                            j += 2
                else:
                    candidate = p[j] # regular case as a candidate.
                    j += 1
                if s[i]==candidate or candidate=='.': # case where match.
                    i += 1
                else:
                    if p[j] == '*':
                        j += 1
                    else:
                        return False

            except IndexError:
                return False
        if j < len(p):
            try:
                if p[-1] != '*':
                    boolean = False
            except IndexError:
                pass;
        return boolean
```
